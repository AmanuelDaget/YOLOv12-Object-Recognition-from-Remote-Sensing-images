```python
import os
import shutil
from pathlib import Path
```

```python
from google.colab import drive
```

```python
mount_point = '/content/drive'
```

```python
ROOT = Path("/content/drive/MyDrive/Colab Notebooks/Computer Vision and Image Processing/Object-detection-using-YOLOv12")

for split in ["training", "test"]:

    split_path = ROOT / split

    images_dir = split_path / "images"
    labels_dir = split_path / "labels"

    # FIX: add parents=True
    images_dir.mkdir(parents=True, exist_ok=True)
    labels_dir.mkdir(parents=True, exist_ok=True)

    for file in split_path.iterdir():

        if file.is_file():

            if file.suffix.lower() in [".jpg", ".jpeg", ".png"]:

                shutil.move(str(file), images_dir / file.name)

            elif file.suffix.lower() == ".txt":

                shutil.move(str(file), labels_dir / file.name)

print("Dataset reorganized successfully.")
```
